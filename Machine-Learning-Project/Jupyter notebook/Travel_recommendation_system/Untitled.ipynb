```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
def recommend(place):
    place_index = new_df[new_df['City'] == place].index[0]
    distances = similarity[place_index]
    places_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

    for i in places_list:
        print(new_df.iloc[i[0]].Place_Name)
    
    
```
